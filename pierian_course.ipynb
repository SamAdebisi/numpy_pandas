```python
def quality(total_bill, tip):
    if tip/total_bill > 0.15:
        return "Generous"
    else:
        return "Other"
```

```python
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
df['Quality 2'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```

```python
df.head()
```

```python
setup = '''
import numpy as np
import pandas as pd

df = pd.read_csv('tips.csv')

def quality(total_bill, tip):
    if tip/total_bill > 0.15:
        return "Generous"
    else:
        return "Other"
'''
```

```python
stmt_one = '''
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
'''
```

```python
stmt_two = '''
df['Quality 2'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
```

```python
import timeit
```

```python
timeit.timeit(setup=setup, stmt=stmt_one, number=1000)
```

```python
timeit.timeit(setup=setup, stmt=stmt_two, number=1000)
```
